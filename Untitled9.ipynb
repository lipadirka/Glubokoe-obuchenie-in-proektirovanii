```python
# Подключ библиотек
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from keras.optimizers import Adam
from keras.utils import to_categorical
from keras.applications import ResNet50
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
# Загрузка dataseta Cifar10
(a_tr, b_tr), (a_t, b_t) = cifar10.load_data()

# Деление пикселей
a_tr = a_tr.astype('float32') / 255
a_t = a_t.astype('float32') / 255
b_tr = to_categorical(b_tr, 12)
b_t = to_categorical(b_t, 12)

# Создание предобученной модели ResNet
resnet = ResNet50(weights='imagenet', include_top=False, input_shape=(32, 32, 3))
resnet.trainable = False

# Создание новой модели
modelres = Sequential()
modelres.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
modelres.add(MaxPooling2D((2, 2)))
modelres.add(Conv2D(32, (3, 3), activation='relu'))
modelres.add(MaxPooling2D((2, 2)))
modelres.add(Conv2D(128, (3, 3), activation='relu'))
modelres.add(MaxPooling2D((2, 2)))
modelres.add(Flatten())
modelres.add(Dense(128, activation='relu'))
modelres.add(Dropout(0.3))
modelres.add(Dense(64, activation='relu'))
modelres.add(Dropout(0.3))
modelres.add(Dense(32, activation='relu'))
modelres.add(Dropout(0.3))
modelres.add(Dense(12, activation='softmax'))

# Компиляц нашей модели
modelres.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Начало обучения модели
history = modelres.fit(a_tr, b_tr, batch_size=2048, epochs=120, validation_data=(a_t, b_t))

# Оценка результата обучения
loss, accuracy = modelres.evaluate(a_t, b_t)
print(f'Ошибка: {loss}, Точность: {accuracy}')
```
