```python
def preprocess_sentence(w):
    """
        Функция для предобработки
    """

    # уменьшаем регистр и убираем лишние пробелы
    w = w.lower().strip()

    # делаем между словом и знаком пунктуации отступ 'привет! -> привет !'
    w = re.sub(r"([?.!,])", r" \1 ", w)
    w = re.sub(r'[" "]+', " ", w)

    # выкидываем все остальные символы из рассмотрения
    w = re.sub(r"[^a-яА-Яa-zA-Z?.!,]+", " ", w)
    w = w.rstrip().strip()

    # добавляем токены для начала и конца предложения
    w = '<start> ' + w + ' <end>'
    return w


def create_dataset(path, num_examples):
    """
        Функция, которая создаёт датасет
    """
    lines = open(path).read().strip().split('\n')

    word_pairs = [[preprocess_sentence(w)
                   for w in l.split('CC-BY 2.0')[0].strip().split('\t')]  for l in lines[:num_examples]]

    return word_pairs
```

```python
class LanguageIndex():
    def __init__(self, lang):
        self.lang = lang
        self.word2idx = {}
        self.idx2word = {}
        self.vocab = set()
        self.create_index()

    def create_index(self):
        for phrase in self.lang:
            self.vocab.update(phrase.split(' '))
        self.vocab = sorted(self.vocab)
        self.word2idx['<pad>'] = 0
        for index, word in enumerate(self.vocab):
            self.word2idx[word] = index + 1
        for word, index in self.word2idx.items():
            self.idx2word[index] = word
```

```python
def get_max_length(tensor):
    return max(len(t) for t in tensor)
```

```python
def load_dataset(path, num_examples):
    pairs = create_dataset(path, num_examples)
    inp_lang_en_ru = LanguageIndex(en for en, ru in pairs)
    targ_lang_en_ru = LanguageIndex(ru for en, ru in pairs)
    inp_lang_ru_en = LanguageIndex(ru for en, ru in pairs)
    targ_lang_ru_en = LanguageIndex(en for en, ru in pairs)

    input_tensor_en_ru = [[inp_lang_en_ru.word2idx[s] for s in en.split(' ')] for en, ru in pairs]
    target_tensor_en_ru = [[targ_lang_en_ru.word2idx[s] for s in ru.split(' ')] for en, ru in pairs]

    input_tensor_ru_en = [[inp_lang_ru_en.word2idx[s] for s in ru.split(' ')] for en, ru in pairs]
    target_tensor_ru_en = [[targ_lang_ru_en.word2idx[s] for s in en.split(' ')] for en, ru in pairs]

    max_length_inp_en_ru = get_max_length(input_tensor_en_ru)
    max_length_tar_en_ru = get_max_length(target_tensor_en_ru)
    max_length_inp_ru_en = get_max_length(input_tensor_ru_en)
    max_length_tar_ru_en = get_max_length(target_tensor_ru_en)

    input_tensor_en_ru = tf.keras.preprocessing.sequence.pad_sequences(input_tensor_en_ru, maxlen=max_length_inp_en_ru, padding='post')
    target_tensor_en_ru = tf.keras.preprocessing.sequence.pad_sequences(target_tensor_en_ru, maxlen=max_length_tar_en_ru, padding='post')

    input_tensor_ru_en = tf.keras.preprocessing.sequence.pad_sequences(input_tensor_ru_en, maxlen=max_length_inp_ru_en, padding='post')
    target_tensor_ru_en = tf.keras.preprocessing.sequence.pad_sequences(target_tensor_ru_en, maxlen=max_length_tar_ru_en, padding='post')

    return (input_tensor_en_ru, target_tensor_en_ru, inp_lang_en_ru, targ_lang_en_ru, max_length_inp_en_ru, max_length_tar_en_ru), \
           (input_tensor_ru_en, target_tensor_ru_en, inp_lang_ru_en, targ_lang_ru_en, max_length_inp_ru_en, max_length_tar_ru_en)
```

```python
# как отдельная функция, чтобы ваше внимание обратить
def gru(units):
    """
        GRU-ячейка
    """
    return tf.keras.layers.GRU(units,
                               return_sequences=True,
                               return_state=True,
                               recurrent_activation='sigmoid',
                               recurrent_initializer='glorot_uniform')
```

```python
class Encoder(tf.keras.Model):
    """
    Наш энкодер
    """

    def __init__(self, vocab_size, embedding_dim, enc_units, batch_sz):
        super(Encoder, self).__init__()
        self.batch_sz = batch_sz
        self.enc_units = enc_units

        # фактически мы будем учит на входе эмбединг
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.enc_units,
                                       return_sequences=True,
                                       return_state=True,
                                       recurrent_initializer='glorot_uniform')

    def call(self, x, hidden):
        x = self.embedding(x)

        # а после к этому эмбедингу применять рекурентность
        output, state = self.gru(x, initial_state = hidden)
        return output, state

    def initialize_hidden_state(self):
        return tf.zeros((self.batch_sz, self.enc_units))
```

```python
class Decoder(tf.keras.Model):
    """
    Наш декодер
    """

    def __init__(self, vocab_size, embedding_dim, dec_units, batch_sz):
        super(Decoder, self).__init__()
        self.batch_sz = batch_sz
        self.dec_units = dec_units
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.dec_units,
                                       return_sequences=True,
                                       return_state=True,
                                       recurrent_initializer='glorot_uniform')
        self.fc = tf.keras.layers.Dense(vocab_size)

        # Веса для внимания
        self.W1 = tf.keras.layers.Dense(self.dec_units)
        self.W2 = tf.keras.layers.Dense(self.dec_units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, x, hidden, enc_output):



        hidden_with_time_axis = tf.expand_dims(hidden, 1)


        score = self.V(tf.nn.tanh(self.W1(enc_output) + self.W2(hidden_with_time_axis)))


        attention_weights = tf.nn.softmax(score, axis=1)


        context_vector = attention_weights * enc_output
        context_vector = tf.reduce_sum(context_vector, axis=1)


        x = self.embedding(x)


        x = tf.concat([tf.expand_dims(context_vector, 1), x], axis=-1)


        output, state = self.gru(x)


        output = tf.reshape(output, (-1, output.shape[2]))


        x = self.fc(output)

        return x, state, attention_weights

    def initialize_hidden_state(self):
        return tf.zeros((self.batch_sz, self.dec_units))
```

```python
# Объединяем две сетки в одну цельную архитектуру
encoder_en_ru = Encoder(len(inp_lang_en_ru.word2idx), embedding_dim, units, BATCH_SIZE)
decoder_en_ru = Decoder(len(targ_lang_en_ru.word2idx), embedding_dim, units, BATCH_SIZE)

encoder_ru_en = Encoder(len(inp_lang_ru_en.word2idx), embedding_dim, units, BATCH_SIZE)
decoder_ru_en = Decoder(len(targ_lang_ru_en.word2idx), embedding_dim, units, BATCH_SIZE)
```

```python
optimizer_en_ru = tf.keras.optimizers.Adam()
optimizer_ru_en = tf.keras.optimizers.Adam()

checkpoint_dir_en_ru = './training_checkpoints_en_ru'
checkpoint_prefix_en_ru = os.path.join(checkpoint_dir_en_ru, "ckpt")

checkpoint_dir_ru_en = './training_checkpoints_ru_en'
checkpoint_prefix_ru_en = os.path.join(checkpoint_dir_ru_en, "ckpt")

checkpoint_en_ru = tf.train.Checkpoint(optimizer=optimizer_en_ru, encoder=encoder_en_ru, decoder=decoder_en_ru)
checkpoint_ru_en = tf.train.Checkpoint(optimizer=optimizer_ru_en, encoder=encoder_ru_en, decoder=decoder_ru_en)

# В качестве функции потерь будет обычная кросс-энтропия на все наши классы (слова)
def loss_function(real, pred):
    mask = tf.math.logical_not(tf.math.equal(real, 0))
    loss_ = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=real, logits=pred)
    mask = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask
    return tf.reduce_mean(loss_)
```

```python
EPOCHS = 5

for epoch in range(EPOCHS):
    start = time.time()

    hidden_en_ru = encoder_en_ru.initialize_hidden_state()
    total_loss_en_ru = 0

    hidden_ru_en = encoder_ru_en.initialize_hidden_state()
    total_loss_ru_en = 0

    for (batch, (inp, targ)) in enumerate(dataset_en_ru):
        loss = 0

        with tf.GradientTape() as tape:
            enc_output, enc_hidden = encoder_en_ru(inp, hidden_en_ru)
            dec_hidden = enc_hidden
            dec_input = tf.expand_dims([targ_lang_en_ru.word2idx['<start>']] * BATCH_SIZE, 1)

            for t in range(1, targ.shape[1]):
                predictions, dec_hidden, _ = decoder_en_ru(dec_input, dec_hidden, enc_output)
                loss += loss_function(targ[:, t], predictions)
                dec_input = tf.expand_dims(targ[:, t], 1)

        batch_loss = (loss / int(targ.shape[1]))
        total_loss_en_ru += batch_loss

        variables = encoder_en_ru.trainable_variables + decoder_en_ru.trainable_variables
        gradients = tape.gradient(loss, variables)
        optimizer_en_ru.apply_gradients(zip(gradients, variables))

        if batch % 100 == 0:
            print('Epoch {} Batch {} Loss {:.4f}'.format(epoch + 1, batch, batch_loss.numpy()))

    if (epoch + 1) % 2 == 0:
        checkpoint_en_ru.save(file_prefix=checkpoint_prefix_en_ru)

    print('Epoch {} Loss {:.4f}'.format(epoch + 1, total_loss_en_ru / N_BATCH))
    print('Time taken for 1 epoch {} sec\n'.format(time.time() - start))

    start = time.time()
    hidden_ru_en = encoder_ru_en.initialize_hidden_state()
    total_loss_ru_en = 0

    for (batch, (inp, targ)) in enumerate(dataset_ru_en):
        loss = 0

        with tf.GradientTape() as tape:
            enc_output, enc_hidden = encoder_ru_en(inp, hidden_ru_en)
            dec_hidden = enc_hidden
            dec_input = tf.expand_dims([targ_lang_ru_en.word2idx['<start>']] * BATCH_SIZE, 1)

            for t in range(1, targ.shape[1]):
                predictions, dec_hidden, _ = decoder_ru_en(dec_input, dec_hidden, enc_output)
                loss += loss_function(targ[:, t], predictions)
                dec_input = tf.expand_dims(targ[:, t], 1)

        batch_loss = (loss / int(targ.shape[1]))
        total_loss_ru_en += batch_loss

        variables = encoder_ru_en.trainable_variables + decoder_ru_en.trainable_variables
        gradients = tape.gradient(loss, variables)
        optimizer_ru_en.apply_gradients(zip(gradients, variables))

        if batch % 100 == 0:
            print('Epoch {} Batch {} Loss {:.4f}'.format(epoch + 1, batch, batch_loss.numpy()))

    if (epoch + 1) % 2 == 0:
        checkpoint_ru_en.save(file_prefix=checkpoint_prefix_ru_en)

    print('Epoch {} Loss {:.4f}'.format(epoch + 1, total_loss_ru_en / N_BATCH))
    print('Time taken for 1 epoch {} sec\n'.format(time.time() - start))
```

```python
def evaluate(sentence, encoder, decoder, inp_lang, targ_lang, max_length_inp, max_length_targ, units):
    attention_plot = np.zeros((max_length_targ, max_length_inp))

    sentence = preprocess_sentence(sentence)

    inputs = []
    for word in sentence.split(' '):
        if word in inp_lang.word2idx:
            inputs.append(inp_lang.word2idx[word])
        else:
            # Обработка слов, отсутствующих в словаре
            raise KeyError(f"Слово '{word}' отсутствует в словаре.")

    inputs = tf.keras.preprocessing.sequence.pad_sequences([inputs], maxlen=max_length_inp, padding='post')
    inputs = tf.convert_to_tensor(inputs)

    result = ''

    hidden = [tf.zeros((1, units))]
    enc_out, enc_hidden = encoder(inputs, hidden)

    dec_hidden = enc_hidden
    dec_input = tf.expand_dims([targ_lang.word2idx['<start>']], 0)

    for t in range(max_length_targ):
        predictions, dec_hidden, attention_weights = decoder(dec_input, dec_hidden, enc_out)

        attention_weights = tf.reshape(attention_weights, (-1, ))
        attention_plot[t] = attention_weights.numpy()

        predicted_id = tf.argmax(predictions[0]).numpy()

        result += targ_lang.idx2word[predicted_id] + ' '

        if targ_lang.idx2word[predicted_id] == '<end>':
            return result, sentence, attention_plot

        dec_input = tf.expand_dims([predicted_id], 0)

    return result, sentence, attention_plot

def plot_attention(attention, sentence, predicted_sentence):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(1, 1, 1)
    ax.matshow(attention, cmap='viridis')

    fontdict = {'fontsize': 14}

    ax.set_xticklabels([''] + sentence, fontdict=fontdict, rotation=90)
    ax.set_yticklabels([''] + predicted_sentence, fontdict=fontdict)

    plt.show()

def translate(sentence, encoder, decoder, inp_lang, targ_lang, max_length_inp, max_length_targ, units):
    result, sentence, attention_plot = evaluate(sentence, encoder, decoder, inp_lang, targ_lang, max_length_inp, max_length_targ, units)

    print('Ввод: {}'.format(sentence))
    print('Предсказано: {}'.format(result))

    attention_plot = attention_plot[:len(result.split(' ')), :len(sentence.split(' '))]
    plot_attention(attention_plot, sentence.split(' '), result.split(' '))

    return result
```

```python
def dtranslate(sentence,
                     encoder_ru_en, decoder_ru_en, inp_lang_ru_en, targ_lang_ru_en, max_length_inp_ru_en, max_length_targ_ru_en, units_ru_en,
                     encoder_en_ru, decoder_en_ru, inp_lang_en_ru, targ_lang_en_ru, max_length_inp_en_ru, max_length_targ_en_ru, units_en_ru):

    intermediate_translation, _, _ = evaluate(sentence, encoder_ru_en, decoder_ru_en, inp_lang_ru_en, targ_lang_ru_en, max_length_inp_ru_en, max_length_targ_ru_en, units_ru_en)


    intermediate_translation = intermediate_translation.replace('<end>', '').strip()


    final_translation, _, _ = evaluate(intermediate_translation, encoder_en_ru, decoder_en_ru, inp_lang_en_ru, targ_lang_en_ru, max_length_inp_en_ru, max_length_targ_en_ru, units_en_ru)

    return final_translation

# Восстановление последней контрольной точки для обоих направлений
checkpoint_ru_en.restore(tf.train.latest_checkpoint(checkpoint_dir_ru_en))
checkpoint_en_ru.restore(tf.train.latest_checkpoint(checkpoint_dir_en_ru))


units_ru_en = 1024
units_en_ru = 1024

ru_en = translate('Ты мне хороший друг!', encoder_ru_en, decoder_ru_en, inp_lang_ru_en, targ_lang_ru_en, max_length_inp_ru_en, max_length_targ_ru_en, units)
print('Перевод (RU на EN): {}'.format(ru_en))


en_ru = translate('You are a good friend!', encoder_en_ru, decoder_en_ru, inp_lang_en_ru, targ_lang_en_ru, max_length_inp_en_ru, max_length_targ_en_ru, units)
print('Перевод (EN на RU): {}'.format(en_ru))
```
